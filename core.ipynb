# Install Jupyter Notebook

Install Jupyter Notebook using pip or conda, ensuring Python is installed.

# Launch Jupyter Notebook

Launch Jupyter Notebook from the command line or terminal.

# Create a New Notebook

Create a new notebook file in the Jupyter interface.

# Rename the Notebook to "core"

Rename the newly created notebook to 'core' using the file menu or by editing the filename.

# Save the Notebook

Save the notebook to persist the changes.
