```python
# Demonstrate computational complexity by timing model training on different data sizes
import time
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

# Load a sample of the data for demonstration
try:
    df = pd.read_csv('../data/diabetic_data_cleaned.csv')
    # Use a subset for faster computation
    df_sample = df.sample(n=min(50000, len(df)), random_state=42)
    
    # Prepare features (simplified)
    feature_cols = ['age_years', 'time_in_hospital', 'num_lab_procedures', 'num_medications', 'number_inpatient']
    X = df_sample[feature_cols].fillna(0)
    y = df_sample['readmitted_30']
    
    sizes = [1000, 5000, 10000, 25000]  # Different sample sizes
    algorithms = {
        'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
        'Random Forest': RandomForestClassifier(n_estimators=10, random_state=42)  # Small n for speed
    }
    
    results = {}
    complexity_params = {
        'n': len(X),  # number of samples
        'd': X.shape[1],  # number of features
        'i': 100,  # number of iterations (logistic regression)
        'm': 10  # number of trees (random forest)
    }
    
    # Calculate and display numerical complexities
    import math
    print("===== COMPUTATIONAL COMPLEXITY (Numerical) =====")
    print(f"\nParameters: n={complexity_params['n']}, d={complexity_params['d']}, i={complexity_params['i']}, m={complexity_params['m']}")
    print(f"\nLogistic Regression: n * d * i = {complexity_params['n']} * {complexity_params['d']} * {complexity_params['i']} = {complexity_params['n'] * complexity_params['d'] * complexity_params['i']:,}")
    print(f"\nRandom Forest: m * n * log(n) = {complexity_params['m']} * {complexity_params['n']} * log({complexity_params['n']}) = {int(complexity_params['m'] * complexity_params['n'] * math.log(complexity_params['n'])):,}")
    print(f"\nGradient Boosting: m * n * log(n) = {complexity_params['m']} * {complexity_params['n']} * log({complexity_params['n']}) = {int(complexity_params['m'] * complexity_params['n'] * math.log(complexity_params['n'])):,}")
    
    # Time actual training
    print("\n===== ACTUAL TRAINING TIMES =====")
    for name, model in algorithms.items():
        times = []
        for n in sizes:
            if n > len(df_sample):
                continue
            X_sub = X[:n]
            y_sub = y[:n]
            
            X_train, X_test, y_train, y_test = train_test_split(X_sub, y_sub, test_size=0.2, random_state=42)
            
            start = time.time()
            model.fit(X_train, y_train)
            end = time.time()
            
            times.append(end - start)
        
        results[name] = times
        print(f"\n{name}:")
        for size, t in zip(sizes[:len(times)], times):
            print(f"  n={size:5d} samples: {t:.4f} seconds")
    
except FileNotFoundError:
    print("Cleaned data file not found. Run Final_Pipeline.ipynb first to generate it.")
```
